```python
import pandas as pd
import numpy as np
import os


# 1. Load data
# (Note: Assuming your csv file and the script are in the same directory.
# If you still get a "file not found" error, replace the filename below with the absolute path)
file_path = 'Data/co2-emissions-emissions-mmtco2.csv'
df = pd.read_csv(file_path)

# 2. Truncate invalid years: keep only Country, Code, and columns from 1980-2023
cols_to_keep = ['Country', 'Code'] + [str(year) for year in range(1980, 2024)]
# Filter out non-existent columns to prevent errors
cols_to_keep = [c for c in cols_to_keep if c in df.columns]

# Filter out all non-3-letter codes (instantly removes regions and WP projection areas)
df = df[df['Code'].str.len() == 3]
df = df[cols_to_keep]

# 3. Convert wide format to long format (Melt)
df_long = pd.melt(
    df,
    id_vars=['Country', 'Code'],
    var_name='Year',
    value_name='CO2_Emission'
)

# 4. Clean outliers and convert data types
# Replace '--' and other non-numeric characters with NaN
df_long['CO2_Emission'] = df_long['CO2_Emission'].replace('--', np.nan)
df_long['CO2_Emission'] = pd.to_numeric(df_long['CO2_Emission'], errors='coerce')

# Convert Year column to integer
df_long['Year'] = df_long['Year'].astype(int)

# 5. Drop rows with NaN emissions (optional)
df_cleaned = df_long.dropna(subset=['CO2_Emission'])

# Print the first 5 rows to check the result
print("First 5 rows of the cleaned data:")
print(df_cleaned.head())

# 6. Export the cleaned data to a new CSV file

# Define the target directory
output_dir = 'Data/cleaned'

# Create the directory if it doesn't already exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    print(f"Created new directory: {output_dir}")

# Combine the directory path and the filename
output_filename = os.path.join(output_dir, 'cleaned_co2_data_1980_2023.csv')

# Save the DataFrame to a CSV file
# index=False prevents saving the row indices (0, 1, 2...) as a column
df_cleaned.to_csv(output_filename, index=False)

print(f"\nSuccess! The cleaned data has been saved to: {output_filename}")
```

```python
# 1. Load GDP data
file_path_gdp = 'Data/gross-domestic-product-gdp-billion-dollars-at-purchasing-power-parities.csv'
df_gdp = pd.read_csv(file_path_gdp)

# 2. Keep only valid 3-letter country codes
df_gdp = df_gdp[df_gdp['Code'].astype(str).str.len() == 3]

# 3. Filter years: 1980-2023
years = [str(y) for y in range(1980, 2024)]
cols_to_keep = ['Country', 'Code'] + [y for y in years if y in df_gdp.columns]
df_gdp = df_gdp[cols_to_keep]

# 4. Convert wide to long format (Melt)
df_gdp_long = pd.melt(
    df_gdp, 
    id_vars=['Country', 'Code'], 
    var_name='Year', 
    value_name='GDP_Billion_PPP'
)

# 5. Handle missing values and type conversion
df_gdp_long['GDP_Billion_PPP'] = df_gdp_long['GDP_Billion_PPP'].replace(['--', 'NA'], np.nan)
df_gdp_long['GDP_Billion_PPP'] = pd.to_numeric(df_gdp_long['GDP_Billion_PPP'], errors='coerce')
df_gdp_long['Year'] = df_gdp_long['Year'].astype(int)

# 6. Drop rows with missing GDP data
df_gdp_cleaned = df_gdp_long.dropna(subset=['GDP_Billion_PPP'])

print("--- First 5 rows of cleaned GDP data ---")
print(df_gdp_cleaned.head())



# ==========================================
# 7. Export the cleaned GDP data to a new CSV file
# ==========================================

# Define the target directory
output_dir = 'Data/cleaned'

# Ensure the directory exists; if not, create it
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    print(f"Created directory: {output_dir}")

# Combine the directory path and the filename
output_filename = os.path.join(output_dir, 'cleaned_gdp_data_1980_2023.csv')

# Save the DataFrame to a CSV file
# index=False prevents saving the row indices (0, 1, 2...) as a column
df_gdp_cleaned.to_csv(output_filename, index=False)

print(f"\n🎉 Success! Cleaned GDP data exported to: {output_filename}")
```

```python

# 1. Load population data
file_path_pop = 'Data/population-population-people-in-thousands.csv'
df_pop = pd.read_csv(file_path_pop)

# 2. Keep only valid 3-letter country codes
df_pop = df_pop[df_pop['Code'].astype(str).str.len() == 3]

# 3. Filter years: 1980-2023
years = [str(y) for y in range(1980, 2024)]
cols_to_keep = ['Country', 'Code'] + [y for y in years if y in df_pop.columns]
df_pop = df_pop[cols_to_keep]

# 4. Convert wide to long format (Melt)
df_pop_long = pd.melt(
    df_pop, 
    id_vars=['Country', 'Code'], 
    var_name='Year', 
    value_name='Population_Thousands'  # Note: Unit is in thousands
)

# 5. Handle missing values and type conversion
# Replace '--' and 'NA' with standard NaN
df_pop_long['Population_Thousands'] = df_pop_long['Population_Thousands'].replace(['--', 'NA'], np.nan)

# Convert to numeric (float) and integer types
df_pop_long['Population_Thousands'] = pd.to_numeric(df_pop_long['Population_Thousands'], errors='coerce')
df_pop_long['Year'] = df_pop_long['Year'].astype(int)

# 6. Drop rows with missing population data
df_pop_cleaned = df_pop_long.dropna(subset=['Population_Thousands'])

print("--- First 5 rows of cleaned population data ---")
print(df_pop_cleaned.head())

# 7. Export the cleaned population data to a new CSV file

# Define the target directory
output_dir = 'Data/cleaned'

# Create the directory if it doesn't already exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    print(f"Created directory: {output_dir}")

# Combine the directory path and the filename
output_filename = os.path.join(output_dir, 'cleaned_pop_data_1980_2023.csv')

# Save the DataFrame to a CSV file
# index=False prevents saving the row indices (0, 1, 2...) as a column
df_pop_cleaned.to_csv(output_filename, index=False)

print(f"\n🎉 Success! Cleaned population data exported to: {output_filename}")
```

```python
# 1. Load the three cleaned CSV files
df_co2 = pd.read_csv('Data/Cleaned/cleaned_co2_data_1980_2023.csv') 
df_pop = pd.read_csv('Data/Cleaned/cleaned_pop_data_1980_2023.csv')
df_gdp = pd.read_csv('Data/Cleaned/cleaned_gdp_data_1980_2023.csv')

# 2. Merge the three tables
print("Merging data tables...")

# Step 1: Merge CO2 and Population
df_master = pd.merge(df_co2, df_pop, on=['Country', 'Code', 'Year'], how='inner')

# Step 2: Merge GDP into the resulting table
df_master = pd.merge(df_master, df_gdp, on=['Country', 'Code', 'Year'], how='inner')

# 3. Calculate core metrics (Feature Engineering)
print("Calculating core decoupling metrics...")

df_master['Per_Capita_CO2_Tonnes'] = (df_master['CO2_Emission'] / df_master['Population_Thousands']) * 1000

df_master['Carbon_Intensity'] = df_master['CO2_Emission'] / df_master['GDP_Billion_PPP']

# 4. Sort and drop rows with missing metric values
df_master = df_master.sort_values(by=['Country', 'Year']).reset_index(drop=True)
df_master = df_master.dropna(subset=['Per_Capita_CO2_Tonnes', 'Carbon_Intensity'])

print("\n--- Ultimate Master Table Preview ---")
print(df_master.head())

# 5. Export
# Define the target directory
output_dir = 'Data/cleaned'

# Ensure the directory exists
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Combine the directory path and the final filename
output_filename = os.path.join(output_dir, 'Master_CO2_Decoupling_Final.csv')

# Save the DataFrame
# index=False prevents saving the row indices (0, 1, 2...) as a column
df_master.to_csv(output_filename, index=False)

print(f"\n🎉 Success! The comprehensive master table has been saved to: {output_filename}")
```
